```python
import random

def get_winner(user, computer):
    if user == computer:
        return "Draw"
    elif (user == 'r' and computer == 's') or \
         (user == 'p' and computer == 'r') or \
         (user == 's' and computer == 'p'):
        return "User wins"
    else:
        return "Computer wins"

# Score tracking
scores = {'User wins': 0, 'Computer wins': 0, 'Draw': 0}
# scores is a dictionary that tracks how many times each result has occurred:
choices = ['r', 'p', 's']
labels = {'r': 'Rock', 'p': 'Paper', 's': 'Scissors'}

# --- Main game loop ---
while True:
    user_choice = input("Choose [r]ock, [p]aper, or [s]cissors (press 'q' to quit): ")

    if user_choice == 'q':
        total_rounds = scores['User wins'] + scores['Computer wins'] + scores['Draw']
        print("\n--- Final Results ---")
        print(f"Total rounds played: {total_rounds}")
        print(f"User wins: {scores['User wins']} times")
        print(f"Computer wins: {scores['Computer wins']} times")
        print(f"Draws: {scores['Draw']} times")
        print("Thanks for playing!")
        break

    if user_choice not in choices:
        print("Invalid input. Please choose r, p, or s.")
        continue

    computer_choice = random.choice(choices)
    print(f"You chose: {labels[user_choice]}")
    print(f"Computer chose: {labels[computer_choice]}")

    result = get_winner(user_choice, computer_choice)
    print(f"Result: {result}")

    scores[result] += 1
# So scores[result] accesses the current count for that result.
# += 1 adds 1 to that count — updating the score.

    total_rounds = scores['User wins'] + scores['Computer wins'] + scores['Draw']
    print(f"Current Score → Rounds: {total_rounds} | User: {scores['User wins']} | Computer: {scores['Computer wins']} | Draws: {scores['Draw']}")
    print("-" * 20)
```
